```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(2,4, figsize=(20,10))

# Categorical features
sns.countplot(x='City/County', data=data, ax=axes[0,0])
sns.countplot(x='House Type', data=data, ax=axes[0,1])
sns.countplot(x='Location', data=data, ax=axes[0,2])
sns.countplot(x='Property Name', data=data, ax=axes[1,0])

# Numerical Features
sns.histplot(x='Price', data=data, ax=axes[1,2])
sns.histplot(x='No. of Bathrooms', data=data, ax=axes[1,3])
```
